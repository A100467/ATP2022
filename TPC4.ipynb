```python
#Cinema = [Sala]
#Sala = (nlugares, Vendidos, filme)
#nlugares = Int
#filme = String 
#Vendidos = [Int]


sala1 = (100, [85,86], "Rain Man")
sala2 = (200, [], "Hannibal")
sala3 = (150, [32,68], "Pretty Woman")
sala4 = (250, [1,46,52], "Whiplash")
cinema1 = [sala1, sala2,sala3,sala4]


#(1)Ver Cartaz
def listar(cinema):                      
    for i in cinema:                     
        nlugares, vendidos, filme = i
        if len(str(filme)) != 0:
            print(filme)
            print("---------------")
    return

#(2)Disponibilidade de cada sala
def listardisponibilidades( cinema ):     
    for i in cinema:
        nlugares, vendidos, filme = i
        print(filme, ",", "Número de lugares disponíveis:",(nlugares - len(vendidos)))    
    return 

#(3)Verificar a disponibilidade de um lugar da sala   
def disponivel(cinema, filmeA, lugar):
    condicoes = False
    disponivel = False
    for i in cinema:
        nlugares, vendidos, filme = i                                 
        if filmeA == filme and lugar < nlugares:
            condicoes = True
            if condicoes == True and lugar in vendidos:
                disponivel = False
                print("Esse lugar não se encontra disponível.")             
            else:                              
                disponivel = True
                print("Esse lugar encontra-se disponível.")
    if condicoes == False:
        print("Não foi possível obter uma conclusao, verificar o filme e o lugar do sala.")
    return disponivel

#(4)Vender Bilhete
def vendebilhete( cinema, filmeA, lugar ):  
    condicao = False
    for i in cinema:
        nlugares, vendidos, filme = i
        if filmeA == filme and lugar < nlugares:
            condicao = True
            vendidos.append(lugar)
            print("Foi possível comprar o bilhete, verifique, vendo se o seu lugar foi adicionado à sala do filme que quer ver.")
    if condicao == False:
        print("Não foi possível comprar um bilhete, verificar se o filme indicado está em exibição e se o lugar se encontra dentro da lotação da sala desse mesmo filme.")
    return cinema

#(5)Inserir Sala
def inserirsala(cinema,sala):
    diferente = True
    (nlugares, [], filmenovo) = sala
    for i in cinema:
        nlugares, vendidos, filme = i
        if filmenovo == filme:
            diferente = False
    if diferente == True:
        cinema.append(sala)
    return cinema

menu =  (""" 
        (1)Ver Cartaz                     
        (2)Disponibilidade de cada sala                    
        (3)Verificar a disponibilidade de um lugar da sala                        
        (4)Vender Bilhete                    
        (5)Inserir Sala                              
        (0)Sair  """)
print(menu)
opcao = int(input("Selecione a opcao do menu "))

while opcao != 0:
    if opcao ==1:
        listar(cinema1)
        print(menu)
        opcao = int(input("Selecione a opcao do menu "))
    elif opcao == 2:
        listardisponibilidades(cinema1)
        print(menu)
        opcao = int(input("Selecione a opcao do menu "))
    elif opcao == 3:
        filmeA = input("Indique o nome do filme que quer ver.")
        lugar = int(input("Escolha o lugar da sala que exibe esse filme. "))
        disponivel(cinema1, filmeA, lugar)
        print(menu)
        opcao = int(input("Selecione a opcao do menu "))
    elif opcao == 4:
        filmeA = input("Indique o nome do filme que quer ver.")
        lugar = int(input("Escolha o lugar da sala que exibe esse filme. "))
        print(vendebilhete( cinema1, filmeA, lugar ))
        print(menu)
        opcao = int(input("Selecione a opcao do menu "))
    elif opcao == 5:
        nlugares = int(input("Escolha o número de lugares para a nova sala."))
        filmenovo = input("Escolha o novo filme que será exibido na sala nova.")
        sala = nlugares, [], filmenovo
        print(inserirsala(cinema1, sala))
        print(menu)
        opcao = int(input("Selecione a opcao do menu "))

if opcao == 0:
    print("Escolheu sair da aplicação, até à próxima!")
```
