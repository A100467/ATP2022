```python
#(1)ler ficheiro de maneira organizada
def lerficheiro(filename):
    file = open(filename, "rt")
    lista = []
    linha1 = file.readline()
    linha1 = linha1[:len(linha1) - 1]
    linhachaves = linha1.split(",")
    for linha in file:
        dicionario = {}
        linha = linha[:len(linha)- 1]
        pessoa = linha.split(",")
        for i in range(len(pessoa)):
            dicionario[linhachaves[i]] = pessoa[i]
        lista.append(dicionario)   
    file.close()
    return lista


#(2)funcao calcula distruicao da doenca por sexo
def distribsexo(filename):
    lista = lerficheiro(filename)
    M = 0                                      
    F = 0                                                
    for dicionario in lista:
        if dicionario["temDoenÃ§a"] == str(1):
            if dicionario["sexo"] == "M":
                M = M + 1                            
            elif dicionario["sexo"] == "F":
                F = F + 1
    sexo = {"M": M , "F": F}  
    return sexo


#(3)distribuição da doença por faixa etária

#encontrar o valor mínimo e máximo da idade
def minmaxIdade(filename):
    lista = lerficheiro(filename)
    minimo = int(lista[0][("idade")])
    maximo = int(lista[0][("idade")])
    for dicionario in lista:
        if int(dicionario["idade"]) < minimo:
            minimo = int(dicionario["idade"])
        if int(dicionario["idade"]) > maximo:
            maximo = int(dicionario["idade"])
    minmax = (minimo, maximo)
    return  minmax

#lista de faixas etarias
def faixasEtarias(filename):
    intervalos = []
    max = minmaxIdade(filename)[1]
    min = minmaxIdade(filename)[0]
    a = min
    b = a + 4
    condicao = False
    while condicao == False:
        intervalo = [a,b]
        intervalos.append(intervalo)
        condicao = (b >= max)
        a = a + 5
        b = a + 4                                                                            
    return intervalos


#distribuição da doença por faixa etária
def distribFaixasEtarias(filename):      
    lista = lerficheiro(filename)          
    intervalos = faixasEtarias(filename)    
    listaFE = []
    for intervalo in intervalos:
        doentes = 0
        for dicionario in lista:   
            if intervalo[1] >= int(dicionario["idade"]) >= intervalo[0]:
                if dicionario["temDoenÃ§a"] == str(1):  
                    doentes += 1
                dictFaixasEtarias = {"faixa etária": intervalo , "temDoença": doentes}
        listaFE.append(dictFaixasEtarias)
    return listaFE



#(4)distribuicao pelos níveis de colesterol

#valores mínimos e máximos de colesterol
def minmaxColesterol(filename):
    lista = lerficheiro(filename)
    minimo = int(lista[0][("colesterol")])
    maximo = int(lista[0][("colesterol")])
    for dicionario in lista:
        if int(dicionario["colesterol"]) < minimo:
            minimo = int(dicionario["colesterol"])
        if int(dicionario["colesterol"]) > maximo:
            maximo = int(dicionario["colesterol"])
    minmax = (minimo, maximo)
    return  minmax

#niveis de colesterol
def niveisColesterol(filename):
    niveis = []
    max = minmaxColesterol(filename)[1]
    min = minmaxColesterol(filename)[0]
    a = min
    b = a + 29
    condicao = False
    while condicao == False:
        intervalo = [a,b]
        niveis.append(intervalo)
        condicao = (b >= max)
        a = a + 30
        b = a + 29                                                                               
    return niveis

#distribuicao pelos níveis de colesterol
def distribNiveisColesterol(filename):          
    lista = lerficheiro(filename)             
    niveis = niveisColesterol(filename)     
    listaNC = []
    for intervalo in niveis:
        doentes = 0
        for dicionario in lista:   
            if intervalo[1] >= int(dicionario["colesterol"]) >= intervalo[0]:
                if dicionario["temDoenÃ§a"] == str(1):  
                    doentes +=  1
                dictNiveisColesterol = {"nível de colesterol": intervalo , "temDoença": doentes}
        listaNC.append(dictNiveisColesterol)
    return listaNC

def menu():
    print("""
        MENU
        (1)Leitura de informação do ficheiro 'myheart.csv'                    
        (2)Distribuicao da doença cardíaca por sexo                  
        (3)Distribuição da doença cardíaca por escalões de idade               
        (4)Distribuição da doença cardíaca por nívies de colesterol                                           
        (0)Sair  """)

#programa final
print("Bem vindo, selecione a opção que deseja executar.")
menu()
opcao = int(input("Escolha a opção desejada."))
while opcao != 0:
    if opcao == 1:
        print(lerficheiro('myheart.csv'))
        menu()
        opcao = input("Escolha a opção desejada.")
    elif opcao == 2:
        print(distribsexo('myheart.csv'))
        menu()
        opcao = input("Escolha a opção desejada.")
    elif opcao == 3:
        print(distribFaixasEtarias('myheart.csv'))
        menu()
        opcao = input("Escolha a opção desejada.")
    elif opcao == 4:
        print(distribNiveisColesterol('myheart.csv'))
        menu()
        opcao = input("Escolha a opção desejada.")

if opcao == 0:
    print("Saiu da aplicação, até à próxima.")
```
